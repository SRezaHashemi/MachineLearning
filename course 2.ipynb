```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities=pd.read_csv('california.csv')
cities.head()
```

```python
latitude=cities['latd']
longitude=cities['longd']
area=cities['area_total_km2']
population=cities['population_total']
```

```python
#plt.figzeize(6,12)
plt.scatter(longitude,latitude,s=area,alpha=0.5,c=population)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('Colifornia')
plt.colorbar()
plt.grid()
```
